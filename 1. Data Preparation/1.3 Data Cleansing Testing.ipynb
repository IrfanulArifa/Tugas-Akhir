```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import VarianceThreshold
```

```python
Testing = pd.read_csv('/Users/irfanularifa/Documents/TA_FIX/0. Dataset/Testing.csv')
Testing
```

```python
y_test = Testing['Label']
y_test
```

```python
X_test = Testing.drop(['Label'], axis = 1)
X_test
```

```python
Selector = VarianceThreshold(threshold=0.5)
X_test = Selector.fit_transform(X_test)
X_test
```

```python
list_of_names = Selector.get_feature_names_out(input_features=None)
```

```python
X_test = pd.DataFrame(X_test)
X_test.columns = list_of_names
X_test
```

```python
X_test.to_csv('/Users/irfanularifa/Documents/TA_FIX/0. Dataset/X_test.csv', index=False)
y_test.to_csv('/Users/irfanularifa/Documents/TA_FIX/0. Dataset/y_test.csv', index=False)
```

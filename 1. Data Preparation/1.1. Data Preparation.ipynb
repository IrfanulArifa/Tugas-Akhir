# Library

```python
import pandas as pd
import numpy as np
```

# Import Dataset 

```python
Smiles = pd.read_csv('/Users/irfanularifa/Documents/TA_FIX/0. Dataset/Smiles.csv')
Dataset = pd.read_csv('/Users/irfanularifa/Documents/TA_FIX/0. Dataset/Dataset.csv', sep=';')
```

```python
Smiles = Smiles.drop(['Name'],axis=1)
Smiles
```

```python
Dataset = Dataset.drop(['IC50 (nM)','pIC50 (M)'], axis=1)
Dataset
```

```python
Dataset = Dataset['pIC50']
Dataset
```

```python
avg = np.average(Dataset)
avg
```

```python
result = pd.concat([Smiles, Dataset.reindex(Smiles.index)], axis=1)
result
```

```python
new_label = Dataset.apply(lambda x: 1 if x>avg else 0)
new_label
```

```python
new_df = result.assign(Label=new_label)
new_df
```

```python
new_df['Label'].value_counts()
```

# Data Split

```python
training = new_df.sample(frac=0.7, random_state=25)
testing = new_df.drop(training.index)
```

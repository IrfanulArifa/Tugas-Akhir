```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import joblib
import matplotlib.pyplot as plt
```

```python
best_estimator_rfc = joblib.load('/Users/irfanularifa/Documents/TA_FIX/0. Dataset/RF_best.pickle')
best_estimator_abc = joblib.load('/Users/irfanularifa/Documents/TA_FIX/0. Dataset/AB_best.pickle')
best_estimator_gbc = joblib.load('/Users/irfanularifa/Documents/TA_FIX/0. Dataset/GB_best.pickle')
```

```python
X_train = pd.read_csv('/Users/irfanularifa/Documents/TA_FIX/0. Dataset/Training.csv')
X_train
```

```python
y_train = X_train['Label']
y_train
```

```python
X_train = X_train.drop(['Label','pIC50'], axis=1)
X_train
```

```python
X_test = pd.read_csv('/Users/irfanularifa/Documents/TA_FIX/0. Dataset/Testing.csv')
X_test
```

```python
y_test = X_test['Label']
y_test
```

```python
X_test = X_test.drop(['Label','pIC50'], axis=1)
X_test
```

```python
best_feature_rfc = pd.read_pickle('/Users/irfanularifa/Documents/TA_FIX/0. Dataset/Best Feature Random Forest.pickle')
best_feature_abc = pd.read_pickle('/Users/irfanularifa/Documents/TA_FIX/0. Dataset/Best Feature Ada Boost.pickle')
best_feature_gbc = pd.read_pickle('/Users/irfanularifa/Documents/TA_FIX/0. Dataset/Best Feature Gradient Boosting.pickle')
```

```python
best_feature_rfc = np.array(best_feature_rfc[0])
best_feature_abc = np.array(best_feature_abc[0])
best_feature_gbc = np.array(best_feature_gbc[0])
```

```python
X_train_rfc = X_train[best_feature_rfc]
X_train_abc = X_train[best_feature_abc]
X_train_gbc = X_train[best_feature_gbc]
```

```python
X_test_rfc = X_test[best_feature_rfc]
X_test_abc = X_test[best_feature_abc]
X_test_gbc = X_test[best_feature_gbc]
```

# Random Forest

```python
X_train_rfc
```

```python
X_test_rfc
```

```python
y_pred_train_rfc = best_estimator_rfc.predict(X_train_rfc)
y_pred_test_rfc = best_estimator_rfc.predict(X_test_rfc)
```

```python
print(f'Random Forest model train score: {accuracy_score(y_train, y_pred_train_rfc)}')
print(f'Random Forest model test score: {accuracy_score(y_test, y_pred_test_rfc)}')
```

```python
print('Random Forest Model Report')
print(classification_report(y_test,y_pred_test_rfc))
print(confusion_matrix(y_train,y_pred_train_rfc))
print(classification_report(y_test,y_pred_test_rfc))
print(confusion_matrix(y_test,y_pred_test_rfc))
```

```python
y_pred_train_abc = best_estimator_abc.predict(X_train_abc)
y_pred_test_abc = best_estimator_abc.predict(X_test_abc)
```

```python
print(f'Ada Boost model train score: {accuracy_score(y_train, y_pred_train_abc)}')
print(f'Ada Boost model test score: {accuracy_score(y_test, y_pred_test_abc)}')
```

```python
print('Ada Boosting Model Report')
print(classification_report(y_train,y_pred_train_abc))
print(confusion_matrix(y_train,y_pred_train_abc))
print(classification_report(y_test,y_pred_test_abc))
print(confusion_matrix(y_test,y_pred_test_abc))
```

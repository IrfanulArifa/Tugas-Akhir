```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
random_forest = pd.read_pickle('/Users/irfanularifa/Documents/TA_FIX/0. Dataset/GA-RFC.pickle')
ada_boost = pd.read_pickle('/Users/irfanularifa/Documents/TA_FIX/0. Dataset/GA-ABC.pickle')
gradient_boost = pd.read_pickle('/Users/irfanularifa/Documents/TA_FIX/0. Dataset/GA-GBC.pickle')
```

```python
random_forest.sort_values('objective_scores', ascending=False)
```

```python
ada_boost.sort_values('objective_scores', ascending=False)
```

```python
gradient_boost.sort_values('objective_scores', ascending=False)
```

```python
plt.rcParams["figure.figsize"] = [15.00, 4.50]
plt.rcParams["figure.autolayout"] = True
f, axes = plt.subplots(1, 3)
sns.histplot(data=random_forest, x="objective_scores", kde=True, bins=5, ax= axes[0], )
sns.histplot(data=ada_boost, x="objective_scores", kde=True, bins=5, ax= axes[1])
sns.histplot(data=gradient_boost, x="objective_scores", kde=True, bins=5, ax= axes[2])
plt.show()
```

```python
plt.rcParams["figure.figsize"] = [15.00, 4.50]
plt.rcParams["figure.autolayout"] = True
f, axes = plt.subplots(1, 3)
sns.barplot(data=random_forest, x="run_idx", y="objective_scores", ax= axes[0])
sns.barplot(data=ada_boost, x="run_idx", y="objective_scores", ax= axes[1])
sns.barplot(data=gradient_boost, x="run_idx", y="objective_scores", ax= axes[2])
plt.show()
```

```python
fig = random_forest.sort_values(by='objective_scores', ascending=False, ignore_index=True)['plot'][0]
dict_ =fig.to_dict()
plt.rcParams["figure.figsize"] = [5.70, 4.50]
plt.rcParams["figure.autolayout"] = True
sns.lineplot(x=dict_['data'][0]['x'], y=dict_['data'][0]['y'])
plt.xlabel("(a)")
plt.ylabel("Objective Score")
```

```python
fig = ada_boost.sort_values(by='objective_scores', ascending=False, ignore_index=True)['plot'][0]
dict_ =fig.to_dict()
plt.rcParams["figure.figsize"] = [5.70, 4.50]
plt.rcParams["figure.autolayout"] = True
sns.lineplot(x=dict_['data'][0]['x'], y=dict_['data'][0]['y'])
plt.xlabel("(b)")
plt.ylabel("Objective Score")
```

```python
fig = gradient_boost.sort_values(by='objective_scores', ascending=False, ignore_index=True)['plot'][0]
dict_ =fig.to_dict()
plt.rcParams["figure.figsize"] = [5.70, 4.50]
plt.rcParams["figure.autolayout"] = True
sns.lineplot(x=dict_['data'][0]['x'], y=dict_['data'][0]['y'])
plt.xlabel("(c)")
plt.ylabel("Objective Score")
```
